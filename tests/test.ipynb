```python
import random
from datetime import date, timedelta
```

```python
start = date.fromisoformat('2020-01-01') 
end = date.fromisoformat('2023-03-01') 
first_day = start + timedelta(random.randint(0, (end - start).days))
last_day = first_day + timedelta(random.randint(0, (end - first_day).days))
```

```python
def get_time(day: str = None, next_week: bool = False) -> date:
    """Returns last Wednesday's date."""
    day = date.fromisoformat(day) if day else date.today()
    offset = (day.weekday() - 2) % 7  # Wednesday = 2
    if offset == 0:
        return day
    return day - timedelta(offset) + next_week * timedelta(7)
```
